```python
import ast

# Lista de las columnas de clasificación
columns_to_process = ['comment1_classification', 'comment2_classification', 'comment3_classification',
                      'comment4_classification', 'comment5_classification', 'comment6_classification',
                      'comment7_classification', 'comment8_classification', 'comment9_classification',
                      'comment10_classification']

# Función para convertir cadenas de texto en listas de diccionarios
def convert_to_dict(row):
    try:
        # Convertir la cadena de texto en una lista de diccionarios
        return ast.literal_eval(row)
    except (ValueError, SyntaxError):
        # Manejar errores en caso de que el formato no sea el esperado
        return [{"label": None, "score": None}]

# Procesar cada columna de la lista
for col in columns_to_process:
    # Convertir las cadenas de texto en listas de diccionarios
    df_reddit_labeled[col] = df_reddit_labeled[col].apply(convert_to_dict)
    
    # Crear nombres para las nuevas columnas de etiquetas y puntuaciones
    label_col = col.replace('_classification', '_label')
    score_col = col.replace('_classification', '_score')
    
    # Extraer etiquetas y puntuaciones
    df_reddit_labeled[label_col] = df_reddit_labeled[col].apply(lambda x: x[0]['label'])
    df_reddit_labeled[score_col] = df_reddit_labeled[col].apply(lambda x: x[0]['score'])

# Verificar los resultados
df_reddit_labeled
```

```python
df = df_reddit_labeled
# Filtrar las columnas que contienen "label" y "score"
label_columns = [col for col in df.columns if 'label' in col]
score_columns = [col for col in df.columns if 'score' in col]

# Crear listas para almacenar los scores positivos y negativos por fila
positive_scores = []
negative_scores = []

# Iterar por cada fila y calcular las medias de positive y negative scores
for i, row in df.iterrows():
    pos_scores = []
    neg_scores = []
    for label_col, score_col in zip(label_columns, score_columns):
        if row[label_col] == 'POSITIVE':
            pos_scores.append(row[score_col])
        elif row[label_col] == 'NEGATIVE':
            neg_scores.append(row[score_col])
    
    # Calcular la media si existen scores, de lo contrario, asignar NaN
    positive_scores.append(sum(pos_scores)/len(pos_scores) if pos_scores else float('nan'))
    negative_scores.append(sum(neg_scores)/len(neg_scores) if neg_scores else float('nan'))

# Añadir las columnas calculadas al DataFrame original
df['mean_positive_score'] = positive_scores
df['mean_negative_score'] = negative_scores

# Mostrar el DataFrame con las nuevas columnas
df[['game_name', 'mean_positive_score', 'mean_negative_score']]
```

```python
import pandas as pd
import requests

df = pd.read_csv('/Users/Jaimecalderon/Desktop/IRONHACK/Proyectos/Proyecto final/Videojuegos/df_reddit_coms.csv')

# Definir la URL de la API y el token de autorización
API_URL = "https://api-inference.huggingface.co/models/facebook/bart-large-mnli"
headers = {"Authorization": f"Bearer {API_TOKEN}"}


def query(payload):
    try:
        response = requests.post(API_URL, headers=headers, json=payload)
        response.raise_for_status()  # Esto lanzará un error para respuestas con códigos de estado 4xx/5xx
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error during the API request: {e}")
        return None

def get_highest_score_label(data):
    if 'scores' in data and 'labels' in data:
        max_index = data['scores'].index(max(data['scores']))
        highest_label = data['labels'][max_index]
        return highest_label
    else:
        print(f"Unexpected response format: {data}")
        return None

# Iterar a través de las columnas de comentarios
for i in range(1, 11):
    column_name = f'comment{i}'
    comments = df[column_name].dropna().tolist()
    
    labels = []
    for comment in comments:
        payload = {
            "inputs": comment,
            "parameters": {"candidate_labels": ["good", "intermediate", "bad"]},
        }
        result = query(payload)
        if result is not None:
            highest_label = get_highest_score_label(result)
            labels.append(highest_label)
        else:
            labels.append(None)  # O puedes elegir agregar algún valor predeterminado

    # Agregar las etiquetas al DataFrame
    df[f'label_{column_name}'] = labels

# Guardar el DataFrame en un nuevo archivo CSV
df.to_csv('/mnt/data/df_reddit_coms_with_labels.csv', index=False)
```

```python
import pandas as pd
import torch
from transformers import pipeline

# Cargar el archivo CSV
df = pd.read_csv('/Users/Jaimecalderon/Desktop/IRONHACK/Proyectos/Proyecto final/Videojuegos/df_reddit_coms_columnas.csv')


# Cargar el modelo localmente
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Definir las etiquetas
candidate_labels = ["good", "intermediate", "bad"]

# Iterar a través de las columnas de comentarios
for i in range(1, 11):
    column_name = f'comment {i}'
    comments = df[column_name].dropna().tolist()
    
    labels = []
    for comment in comments:
        result = classifier(comment, candidate_labels)
        highest_label = result['labels'][0]  # La etiqueta con mayor puntuación
        labels.append(highest_label)

    # Agregar las etiquetas al DataFrame
    df[f'label_{column_name}'] = labels

# Guardar el DataFrame en un nuevo archivo CSV
df.to_csv('df_reddit_coms_with_labels_local2.csv', index=False)
```

```python
import pandas as pd

# Verificar las columnas del DataFrame
df.columns

# Contar el total de comentarios y el número de veces que cada etiqueta aparece para cada juego
def calculate_label_percentages(group):
    total_comments = len(group)
    good_count = (group == 'good').sum()
    intermediate_count = (group == 'intermediate').sum()
    bad_count = (group == 'bad').sum()

    return pd.Series({
        'percentage_good': good_count / total_comments * 100,
        'percentage_intermediate': intermediate_count / total_comments * 100,
        'percentage_bad': bad_count / total_comments * 100
    })

# Agrupar por 'game_name' y calcular los porcentajes
label_columns = [f'label_comment{i}' for i in range(1, 11)]
df_percentages = df.groupby('game_name')[label_columns].apply(lambda x: x.apply(calculate_label_percentages)).reset_index()

# Ver las primeras filas del DataFrame resultante
df_percentages
```

```python
# Reformatear el DataFrame para obtener una columna por cada porcentaje en lugar de múltiples filas
df_grouped = df_percentages.pivot(index='game_name', columns='level_1').reset_index()

# Renombrar las columnas para hacerlas más claras
df_grouped.columns = ['game_name'] + [f'{col[1]}_{col[0]}' for col in df_grouped.columns if col[0] != 'game_name']

# Calcular los promedios de los porcentajes para todas las columnas de comentarios
df_grouped['average_percentage_good'] = df_grouped.filter(like='percentage_good').mean(axis=1)
df_grouped['average_percentage_intermediate'] = df_grouped.filter(like='percentage_intermediate').mean(axis=1)
df_grouped['average_percentage_bad'] = df_grouped.filter(like='percentage_bad').mean(axis=1)

# Seleccionar solo las columnas relevantes para el resultado final
df_final = df_grouped[['game_name', 'average_percentage_good', 'average_percentage_intermediate', 'average_percentage_bad']]

# Ver las primeras filas del DataFrame final
df_final.head()

# Guardar el DataFrame resultante en un nuevo archivo CSV

df_final.to_csv('reddit_g_b_i.csv', index=False)
```
