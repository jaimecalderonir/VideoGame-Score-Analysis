```python
import ast
import pandas as pd

# FunciÃ³n para convertir las cadenas en listas y manejar posibles errores
def convert_to_list(numbers):
    # Verifica si 'numbers' es una cadena y trata de convertirla a una lista
    if isinstance(numbers, str):
        try:
            # Convierte la cadena en una lista
            return ast.literal_eval(numbers)
        except (ValueError, SyntaxError) as e:
            print(f"Error al convertir: {numbers} - {e}")
            return []  # Devuelve una lista vacÃ­a si hay un error
    elif isinstance(numbers, list):
        return numbers  # Devuelve la lista tal como estÃ¡ si ya es una lista
    return []  # Devuelve lista vacÃ­a si no es una cadena o lista

# Aplicar la funciÃ³n a la columna 'games_counts' para convertirla a listas
df_info_clean['games_counts'] = df_info_clean['games_counts'].apply(convert_to_list)

# FunciÃ³n para sumar los nÃºmeros en la lista, asegurando que sean enteros o flotantes
def safe_sum(numbers):
    # Suma solo los elementos que son enteros o flotantes
    valid_numbers = [n for n in numbers if isinstance(n, (int, float))]
    if not valid_numbers:
        print(f"No se encontraron nÃºmeros vÃ¡lidos en: {numbers}")
    return sum(valid_numbers)

# Crear la nueva columna 'total_platform_game_counts' con la suma de cada lista
df_info_clean['total_platform_game_counts'] = df_info_clean['games_counts'].apply(safe_sum)

# Mostrar el DataFrame con las columnas relevantes
print(df_info_clean[['games_counts', 'total_platform_game_counts']])
```
