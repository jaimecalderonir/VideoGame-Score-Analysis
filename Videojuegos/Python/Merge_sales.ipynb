```python
df_data_ws.rename(columns={'Title' : 'game_name'},inplace=True)
```

```python
df_consolidado = df_data_ws.groupby('game_name').agg({
    'NA Sales': 'sum',
    'Japan Sales': 'sum',
    'Other Sales': 'sum',
    'Total Sales': 'sum'
}).reset_index()
```

```python
df_clean_final = pd.read_csv('/Users/Jaimecalderon/Desktop/IRONHACK/Proyectos/Proyecto final/Videojuegos/df_clean_final.csv')
```

```python
df1 = df_clean_final
df2 = df_consolidado
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()


sales_columns = ['NA Sales', 'Japan Sales', 'Other Sales', 'Total Sales']

for col in sales_columns:
    df2[col] = pd.to_numeric(df2[col].str.replace(r'[^\d.]', '', regex=True), errors='coerce')

df_merged = pd.merge(df1, df2, on='game_name', how='left')
df_merged
```
