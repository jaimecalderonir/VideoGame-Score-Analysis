```python
df_merge_todo.drop(columns=['Release Date', 'Last Update', 'Year'],inplace=True)
```

```python
df_merge_todo['released'] = pd.to_datetime(df_merge_todo['released'])
```

```python
df_merge_todo['Year'] = df_merge_todo['released'].dt.year
```

```python
df_merge_todo['Year'] == 2015 & df_merge_todo['Year'] == 2024
```

```python
df_con_todo_15_24_1 = df_merge_todo[(df_merge_todo['Year'] >= 2015) & (df_merge_todo['Year'] <= 2024)]
```

```python
df_con_todo_15_24_1 = df_con_todo_15_24_1.fillna('N/A')
```

```python
df_con_todo_15_24_1['name'] = df_con_todo_15_24_1['name'].drop_duplicates()
```

```python
df_con_todo_15_24_1.dropna()
```

```python
df_con_todo_15_24_1.to_csv('df_info_merge_data_vg_ws_clean.csv',index=False)
```

```python
def convertir_a_millones(valor):
    if isinstance(valor, str) and 'm' in valor:
        return int(float(valor.replace('m', '')) * 1_000_000)
    else:
        return None

cols_to_convert = ['Total Sales', 'NA Sales', 'PAL Sales', 'Japan Sales', 'Other Sales']
for col in cols_to_convert:
    df_con_todo_15_24_1[col] = df_con_todo_15_24_1[col].apply(convertir_a_millones)
```
